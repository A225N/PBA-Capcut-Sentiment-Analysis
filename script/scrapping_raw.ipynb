```python
from google_play_scraper import reviews_all, Sort, reviews

import pandas as pd
import time 

APP_ID = "com.lemon.lvoverseas"
OUTPUT_CSV = "../csv/capcut_reviews_raw.csv"
```

```python
def data_scrap (app_id: str, langs: list[str], country: str = "id") -> pd.DataFrame:

    all_dfs=[]
    for lang in langs:
        print(f"Bahasa dari data = {lang}, diambil dari negara {country}")
        start = time.time() 
        result, continuation_token = reviews(
            APP_ID,
            lang=lang,
            country=country,
            sort=Sort.NEWEST,
            count=5000,
        )
        elapsed = time.time() - start
        print(f"waktu scrap{elapsed:.2f}")
        print(f"Data yang terambil: {len(result)}")


        df_lang = pd.DataFrame(result)
        df_lang["lang"] = lang 
        all_dfs.append(df_lang) 


    df_all = pd.concat(all_dfs, ignore_index=True) 
    print(f"Total gabungan semua bahasa: {len(df_all)} review")
    return df_all
```

```python
df = data_scrap (
    app_id = APP_ID,
    langs = ["id"],
    country = "id"
)

df.to_csv(OUTPUT_CSV, index=False)
```
